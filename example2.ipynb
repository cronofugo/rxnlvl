```python
# Importar el módulo
from irxnlvl import *
```

```python
# Crear un objeto diagrama
p = plot(10.0, zero=energy(0.0, 'kjmol'), bgcolour=0xfff, units='kcalmol', digits=1, qualified=True)

# Agregar la línea base
p + baseline(colour=0x0, mode='dashed', opacity=0.1)

# Agregar los niveles de energía
p + level(energy(    0, 'kjmol'),  1,    '1',      0x0)
p + level(energy(  -75.5, 'kjmol'),  2,  'EC1',      0x0)
p + level(energy(   74, 'kjmol'),  3, 'TS1a', 0xFF4444)
p + level(energy(   10, 'kjmol'),  3, 'TS1b',      0x0)
p + level(energy( -102, 'kjmol'),  4,  'DC1',      0x0)
p + level(energy(  -42, 'kjmol'),  5,  'TS2',      0x0)
p + level(energy( -111, 'kjmol'),  6,  'EC2',      0x0)
p + level(energy(  -83, 'kjmol'),  7, 'TS3b',      0x0)
p + level(energy(   18, 'kjmol'),  7, 'TS3a', 0x44FF44)
p + level(energy( -103, 'kjmol'),  8,  'DC2',      0x0)
p + level(energy( -127, 'kjmol'),  9,    '3',      0x0)

# Unir los niveles de energía entre sí
p + edge(    '1',  'EC1', 0x0, 0.4, 'normal')
p + edge(  'EC1', 'TS1a', 0x0, 0.2, 'normal')
p + edge(  'EC1', 'TS1b', 0x0, 0.4, 'normal')
p + edge( 'TS1a',  'DC1', 0x0, 0.2, 'normal')
p + edge( 'TS1b',  'DC1', 0x0, 0.4, 'normal')
p + edge(  'DC1',  'TS2', 0x0, 0.4, 'normal')
p + edge(  'TS2',  'EC2', 0x0, 0.4, 'normal')
p + edge(  'EC2', 'TS3b', 0x0, 0.4, 'normal')
p + edge(  'EC2', 'TS3a', 0x0, 0.2, 'normal')
p + edge( 'TS3a',  'DC2', 0x0, 0.2, 'normal')
p + edge( 'TS3b',  'DC2', 0x0, 0.4, 'normal')
p + edge(  'DC2',    '3', 0x0, 0.4, 'normal')

# Visualizar el diagrama
p.render()
```

```python
# Guardar el diagrama en un archivo
p.write("diagram2.png", scale=2)
```
